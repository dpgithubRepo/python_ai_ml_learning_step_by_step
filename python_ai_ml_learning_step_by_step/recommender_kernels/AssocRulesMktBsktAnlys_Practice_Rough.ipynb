```python
basket =  online_df.groupby(['InvoiceNo','Description'])['Quantity'].sum()
```

```python
basket
```

```python
basket  = basket.unstack().reset_index()
```

```python
basket = basket.fillna(0).set_index("InvoiceNo")
```

```python
basket.sample(10)
```

```python
def encode_quantity(quantity):
    if(quantity>0):
        return 1
    else:
        return 0
```

```python
basket_set = basket.applymap(encode_quantity)
```

```python
basket_set
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
frequent_itemsets = apriori(basket_set,min_support=0.02, use_colnames=True)
```

```python
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets,metric='lift',min_threshold=1)
```

```python
rules.sample(5)
```

```python
rules.sort_values('confidence',ascending=False)
```
