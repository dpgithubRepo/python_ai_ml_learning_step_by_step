# Recommender Systems

Recommender Systems are popular for recommending the users with product like movies, music, news, groceries, books etc and has become more important aspect for cross selling across the industries. Also the Recommender Sytems has the ability to personalize the customers to each individual customer/buyer.

__Algorithms__ : The following are three important Alogirthms prominently used

1. Association Rules
2. Collaborative Filtering
3. Matrix Factorization


__1. Association Rules__ : This is mainly meant to identify the combination of items, which are frequently brought together. This is basically used in MBA (Market Basket Analysis) by retailers to find association between the items. 

For example if a customer buys a beer, the system has to predict/identify what the user may buy along with beer and recomend to him or place the predicted product near to the beer rack. 

With retailers selling millions of records the assocition rule computation becomes intense and requires high computation power. One way of handling this is by  filtering out the items which cannot be part of itemset (basket).The apriori algorithm is used by the association rules to filter out the items. The apriori generates rules like

{diaper}  -> {beer}

which means customers who bought diaper also bought beer.

here diaper is __antecedent__ and beer is __consequent__

{diaper,beer} -> {bread,sugar}

The above is also a valid rule generated by apriori




To predict this association there are various metrics involved in this algorithm.

__a. Support__

__b. Confidence__

__c.Lift__





__Support__: Support indicates the frequency of items appearing together in basket with respect to all the possible baskets considered

Consider the following baskets

1. diaper, beer, sugar
2. diaper, cereal, milk
3. beer, diaper, bread
4. cereal,beerm sugar

N - total number of baskets

N(x,y) - number of baskets in which x,y appear together

N(x) - number of baskets in which x appears

N(y) - number of baskets in which y appears


   __Support(x,y)   =  N(x,y)/N__
   
   In our above example
   
   support(beer,diaper) = 2/4 = 0.5 or 50%
   
   
we can use minimum support to filter out the stronger relation/associated combinations. By setting min support of 0.05 we are stating that consider the itemset(combination of items) which are atleast 5% in total of the baskets

Support value helps selecting the combinations which are worth enough for further investigation
 

__Confidence :__ It is the conditional probability of y given x

P(y|x) =  N(x,y)/N(x)

Confidence meaures the proportion of transactions that contain x, which also contain y.

x - antecedent

y - consequent


__Lift :__ Lift is degree of association between the items

                 lift = support(x,y) / (support(x) * support(y))   = N(x,y)/ (N(x)*N(y))
                 
                 
lift = 1 indicates the items are independent meaning no association

lift < 1 indicates the items are substitutions, meaning purchase of one item will decrease other

lift > 1 induicates the items are associated meaning purchase of one item  will increase other
