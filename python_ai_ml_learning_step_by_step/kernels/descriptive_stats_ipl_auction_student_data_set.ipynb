# Let us use Simple Student data frame to explore more on merge, groupby, lambda function

```python
student_academic_df = pd.DataFrame({"id":[1,2,3,4,5,6,7,8],
                          "name":["durga","prasad","wilian","johny","ranger","joe","martin","rohit"],
                          "statistics":[99,98,87,96,88,78,88,89],
                         "science":[98,88,87,96,88,68,66,79],
                          "english":[99,78,77,68,85,78,88,89]})
```

```python
student_academic_df
```

# lambda function to calculate avg marks

```python
student_academic_df['avg'] = student_academic_df.apply((lambda rec: (rec.english+rec.statistics+rec.science)/3), axis=1)
```

```python
student_academic_df
```

```python
students_sports_df = pd.DataFrame({"id":[1,2,3,4,5,6,7,8],
                          "name":["durga","prasad","wilian","johny","ranger","joe","martin","rohit"],
                          "points":[99,98,87,96,88,78,88,89]})
```

```python
students_sports_df
```

```python
student_academic_df
```

```python
student_over_all_grades = student_academic_df.merge(students_sports_df,on=['id'], how="outer")
```
